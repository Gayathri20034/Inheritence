```python
# Step 1: Import Required Libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# Step 2: Data Loader Class
class DataLoader:
    def __init__(self):
        self.X, self.y = load_iris(return_X_y=True)

    def split(self, test_size=0.3, random_state=42):
        return train_test_split(self.X, self.y, test_size=test_size, random_state=random_state)

# Step 3: Preprocessor Class
class Preprocessor:
    def __init__(self):
        self.scaler = StandardScaler()

    def fit_transform(self, X_train):
        return self.scaler.fit_transform(X_train)

    def transform(self, X_test):
        return self.scaler.transform(X_test)

# Step 4: Base ML Model Class
class MLModel:
    def __init__(self):
        self.model = None  # This will be defined in child class

    def train(self, X_train, y_train):
        self.model.fit(X_train, y_train)

    def predict(self, X_test):
        return self.model.predict(X_test)

# Step 4.1: Subclass - Decision Tree
class DecisionTreeModel(MLModel):
    def __init__(self):
        super().__init__()
        self.model = DecisionTreeClassifier()

# Step 4.2: Subclass - Random Forest
class RandomForestModel(MLModel):
    def __init__(self):
        super().__init__()
        self.model = RandomForestClassifier()

# Step 4.3: Subclass - Logistic Regression
class LogisticRegressionModel(MLModel):
    def __init__(self):
        super().__init__()
        self.model = LogisticRegression(max_iter=200)

# Step 5: Evaluator Class
class Evaluator:
    def __init__(self, y_true, y_pred):
        self.y_true = y_true
        self.y_pred = y_pred

    def report(self):
        print("ðŸ“Š Classification Report:\n")
        print(classification_report(self.y_true, self.y_pred))

# Step 6: Main ML Application Class
class MLApplication:
    def __init__(self, model_type):
        self.loader = DataLoader()
        self.preprocessor = Preprocessor()

        # Choose model type using inheritance
        if model_type == "decision_tree":
            self.model = DecisionTreeModel()
        elif model_type == "random_forest":
            self.model = RandomForestModel()
        elif model_type == "logistic_regression":
            self.model = LogisticRegressionModel()
        else:
            raise ValueError("Unsupported model type!")

    def run(self):
        # Load and split data
        X_train, X_test, y_train, y_test = self.loader.split()

        # Preprocess data
        X_train_scaled = self.preprocessor.fit_transform(X_train)
        X_test_scaled = self.preprocessor.transform(X_test)

        # Train model
        self.model.train(X_train_scaled, y_train)

        # Predict
        y_pred = self.model.predict(X_test_scaled)

        # Evaluate
        evaluator = Evaluator(y_test, y_pred)
        evaluator.report()

# Step 7: Execute the Pipeline
app = MLApplication(model_type="random_forest")  # Options: decision_tree, random_forest, logistic_regression
app.run()
```
